```python
#Q3. Write a Pandas program to add, subtract, multiple and divide two Pandas Series.
# Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 9]
```

```python
def function3():
    n1=pd.Series([2,4,6,8,10])
    n2=pd.Series([1,3,5,7,9])
    print(f" N1={n1}")
    print(f"N2={n2}")
    print("*"*50)
    print(n1+n2)
    print("*"*50)
    print(n1-n2)
    print("*"*50)
    print(n1*n2)
    print("*"*50)
    print(n1/n2)
    print("*"*50)



    
    
```

```python
function3()
```

```python
#Q6. Write a Pandas program to change the data type of given a column or a Series.
```

```python
def function6():
    s1= pd.Series(['100','200','Nikhil','300.15','400'])
    print("Original series:")
    print(s1)
    print("change the data type:")
    s2=pd.to_numeric(s1,errors='coerce')
    print(s2)
```

```python
function6()
```
